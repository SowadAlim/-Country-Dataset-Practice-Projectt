# Importing Libraries and the Data frame

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df=pd.read_csv("survey_results_public.csv")

df
```

# Singling out the "Country" column

```python
df1= df[["Country"]]
df1.head(100)
```

# Attempting to find missing values

```python
df2 =(df1.isnull().sum())
df2
```

# Attempting to drop "N/A's"

```python
df3 = df2.dropna()
df3
```

# Attempting to reset the index

```python
df3 = df3.reset_index(drop = True)
df3
```

# Counting the different countries in the data frame

```python
df2= df1.groupby('Country').count()
df2
```

# Plotting the many different countries and the number of the particpants from the countries

```python
df1['Country'].value_counts().plot(kind="bar", figsize=(15,7))
column_labels = list('ABCD')
row_labels = list('WXYZ')
data = np.random.rand(4, 4)
fig, ax = plt.subplots()
heatmap = ax.pcolor(data, cmap=plt.cm.Blues)

# put the major ticks at the middle of each cell
ax.set_xticks(np.arange(data.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(data.shape[0]) + 0.5, minor=False)

# want a more natural, table-like display
ax.invert_yaxis()
ax.xaxis.tick_top()

ax.set_xticklabels(column_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

plt.xlabel("Variety of Countries")
plt.ylabel("Frequency Number")
plt.title("Country")
plt.show()
```
